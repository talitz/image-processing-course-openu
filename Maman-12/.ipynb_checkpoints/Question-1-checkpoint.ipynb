```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# loads an input image in grayscale mode
grayScaleInImage = mpimg.imread('cameraman.jpg', cv2.IMREAD_GRAYSCALE)

inImageArray = np.array(grayScaleInImage)
# creates a figure and two subplots with one row and two columns

figure, (plotBeforeFilter, plotAfterFilter) = plt.subplots(1, 2)

# sets the titles
plotBeforeFilter.set_title(' Before running the filter ')
plotAfterFilter.set_title(' After running the filter ')

# displays the plot before running the filer with gray colormap
plotBeforeFilter.imshow(inImageArray, cmap='gray')

# defines the filter
filter = np.array([[ 1 , 1 , 1 ],
[ 1 , -8 , 1 ],
[ 1 , 1 , 1 ]])

# builds a 2D filter
imageAfterFilter = cv2.filter2D(inImageArray, -1 , filter )
''' saves the result image after the filter
(it is not part of the question, so I run
it only for giving me the output result image
in order to attach it to my zip file)'''

#cv2.imwrite('imageAfterFilter.jpg', imageAfterFilter)

# displays the image result with gray colormap
plotAfterFilter.imshow(imageAfterFilter, cmap=GRAY_COLORMAP)
# displays the figure
plt.show()
```
