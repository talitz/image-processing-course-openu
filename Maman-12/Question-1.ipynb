```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

grayScaleInImage = mpimg.imread('cameraman.jpg', cv2.IMREAD_GRAYSCALE)

inImageArray = np.array(grayScaleInImage)

figure, (plotBeforeFilter, plotAfterFilter) = plt.subplots(1, 2)

plotBeforeFilter.set_title('Before Filtering')
plotAfterFilter.set_title('After Filtering')
plotBeforeFilter.imshow(inImageArray, cmap='gray')

filter = np.array([[ -1 , -2 , -1 ],[ 0 , 0 , 0 ],[ 1 , 2 , 1 ]])

imageAfterFilter = cv2.filter2D(inImageArray, -1 , filter )

cv2.imwrite('imageAfterFilter.jpg', imageAfterFilter)
plotAfterFilter.imshow(imageAfterFilter, cmap='gray')
plt.show()
```
